```python
x = np.arange(-10,10,0.02)
def ReLU(x):
    return x * (x > 0)

def leaky_ReLU(x):
    y1 = ((x > 0) * x)                                                 
    y2 = ((x <= 0) * x * 0.07)                                         
    leaky_way2 = y1 + y2  
    return leaky_way2
```
